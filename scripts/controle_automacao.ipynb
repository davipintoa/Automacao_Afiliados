```python
# Enviar mensagem de promoção no WhatsApp Web com imagem e texto formatado
# Autor: Davi Almeida
# Versão ajustada para ler os dados da automação meli

import subprocess
import time
import pyautogui
import json
from PIL import Image
import win32clipboard
import io



pyautogui.click(1702, 1054)  # Clica fora para garantir que a janela correta está ativa
time.sleep(1)  # Tempo para você posicionar o mouse
pyautogui.click(1686, 600)  # Clica fora para garantir que a janela correta está ativa


# -------------------------------------------------------------------
# Função para copiar imagem para área de transferência
# -------------------------------------------------------------------
def copiar_imagem(imagem_path):
    try:
        image = Image.open(imagem_path)
        output = io.BytesIO()
        image.convert("RGB").save(output, "BMP")
        data = output.getvalue()[14:]  # remove cabeçalho BMP
        output.close()

        win32clipboard.OpenClipboard()
        win32clipboard.EmptyClipboard()
        win32clipboard.SetClipboardData(win32clipboard.CF_DIB, data)
        win32clipboard.CloseClipboard()
        return True
    except Exception as e:
        print(f"Erro ao copiar imagem: {e}")
        return False

# -------------------------------------------------------------------
# Função para copiar texto Unicode ao clipboard
# -------------------------------------------------------------------
def copiar_texto(texto):
    win32clipboard.OpenClipboard()
    win32clipboard.EmptyClipboard()
    win32clipboard.SetClipboardData(win32clipboard.CF_UNICODETEXT, texto)
    win32clipboard.CloseClipboard()

# -------------------------------------------------------------------
# Caminhos e configurações
# -------------------------------------------------------------------
chrome_path  = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
profile_name = "Profile 2"
whatsapp_url = "https://web.whatsapp.com"
grupo        = "Achei no Outlet"  # Nome exato do grupo ou contato

# CAMINHO AJUSTADO: Aponta para o arquivo JSON correto gerado pela Célula 1
mensagens_json = r"C:\Users\davi_\OneDrive\Documentos\Trabalho\Achei no Outlet\meus_scripts\Automacao_Afiliados\output\mensagens_json\mensagens_meli.json"


# -------------------------------------------------------------------
# Início da Automação
# -------------------------------------------------------------------

# Abre o WhatsApp Web no Chrome com o perfil padrão
print("Abrindo o WhatsApp Web...")
subprocess.Popen([
    chrome_path,
    f"--profile-directory={profile_name}",
    whatsapp_url
])
print("Aguardando 15 segundos para o WhatsApp carregar...")
time.sleep(15)

# Seleciona o grupo usando o atalho
print(f"Buscando o grupo '{grupo}'...")
pyautogui.hotkey("ctrl", "alt", "/")
time.sleep(1)
pyautogui.typewrite(grupo)
time.sleep(2)
pyautogui.press("enter")
time.sleep(2)

# Carrega a lista de promoções do arquivo JSON
print(f"Carregando promoções do arquivo: {mensagens_json}")
try:
    with open(mensagens_json, "r", encoding="utf-8") as f:
        produtos = json.load(f)
    print(f"{len(produtos)} promoções encontradas.")
except FileNotFoundError:
    print("ERRO: Arquivo JSON não encontrado. Execute a Célula 1 primeiro.")
    produtos = []
except json.JSONDecodeError:
    print("ERRO: O arquivo JSON está mal formatado ou corrompido.")
    produtos = []

# Envia cada promoção
for item in produtos:
    nome_produto = item.get("nome", "Produto")
    print(f"Enviando: {nome_produto}")
    mensagem    = item.get("mensagem", "")
    # Pega o caminho da imagem diretamente do JSON
    imagem_path = item.get("imagem", "")

    # Verifica se o caminho da imagem existe e é válido
    if imagem_path and os.path.exists(imagem_path):
        # --- LÓGICA PARA ENVIAR IMAGEM COM LEGENDA ---
        if copiar_imagem(imagem_path):
            print("  - Imagem copiada. Colando no WhatsApp...")
            # 1. Cola a imagem
            pyautogui.hotkey("ctrl", "v")
            time.sleep(3)  # Espera a tela de pré-visualização da imagem
            
            # 2. Copia e cola a mensagem como legenda
            print("  - Adicionando legenda...")
            copiar_texto(mensagem)
            pyautogui.hotkey("ctrl", "v")
            time.sleep(1)

            # 3. Envia a imagem com a legenda
            pyautogui.press("enter")
        else:
             print("  - Erro ao copiar imagem. Enviando apenas texto...")
             copiar_texto(mensagem)
             pyautogui.hotkey("ctrl", "v")
             time.sleep(1)
             pyautogui.press("enter")
    
    elif mensagem:
        # --- LÓGICA PARA ENVIAR APENAS TEXTO ---
        print("  - Imagem não encontrada. Enviando apenas texto...")
        copiar_texto(mensagem)
        pyautogui.hotkey("ctrl", "v")
        time.sleep(1)
        pyautogui.press("enter")

    print("Mensagem enviada. Aguardando 5 segundos...")
    time.sleep(5)

print("\n\n✅ Processo de envio finalizado!")
```
